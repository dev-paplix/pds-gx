```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Basic Plot

plt.plot([1,2,3], [2,4,6])
plt.show()
```

```python
# Basic plot with x and y

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
plt.plot(x,y)
plt.show()
```

```python
# Basic plot but some visual changes

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
plt.plot(x,y, label='2X', color='#6e11d9', linewidth=5, markersize=10, markeredgecolor='red')
plt.show()
```

```python
# Basic plot but some markers

plt.figure(figsize = (5,3), dpi=300)
x2 = np.arange(0,4.5,0.5)
y2 = x2**3
plt.plot(x2, y2, label='3x', color='blue', linewidth=2, marker='p', markersize=10, markeredgecolor='red')
plt.title(
    'This is my first marked chart', fontdict={'fontname': 'monospace', 'fontsize':20}
)
plt.xlabel('X label:', fontdict={'fontname': 'monospace', 'fontsize':15} )
plt.ylabel('Y label:', fontdict={'fontname': 'monospace', 'fontsize':15} )
plt.xticks([0, 1, 3, 5])
plt.yticks([20,40,60,80])
plt.legend(['2X', '3X'], loc = 'upper left')
plt.savefig('./output/first_plot.png', dpi = 300)
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
plt.bar(labels, values)
plt.show()
```

```python
species = ("Adelie", "Chinstrap", "Gentoo")
penguins_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout='constrained')
for attribute, measurment in penguins_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurment, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier +=1

ax.set_ylabel('Length (mm)')
ax.set_title('Penguins attributes by species')
ax.set_xticks(x+ width, species)
ax.legend(loc = 'upper left', ncols=3)
ax.set_ylim(0,250)
plt.show()
```

```python
commodity_prices = pd.read_csv("./assets/commodity_prices.csv")
commodity_prices.tail(5)
```

```python
# Visualizing Two Major oil supply fluctuation over time

plt.figure(figsize=(15,6), dpi=300)
plt.plot(commodity_prices['date'], commodity_prices['oil_brent'], label='Brent Oil', color= "green")
plt.plot(commodity_prices['date'], commodity_prices['oil_dubai'], label='Dubai Oil', color= "black")
plt.title('Oil Price in Last 62 years')
plt.xlabel('Date')
plt.ylabel('Price in $')
plt.legend(loc='upper left')
plt.xticks([100,150 ,200, 250,300,350, 400,450, 500, 550,600,650, 700,750, 800,850, 900], rotation=45)
plt.show()
```

```python
commodity_prices.tail(4)
```

```python
# Showing the average price of coffee over time

avg_prices = commodity_prices[['coffee_arabica', 'coffee_robustas']].mean()
plt.figure(figsize= (8,3), dpi=300)
plt.bar(avg_prices.index, avg_prices.values, color=['brown', 'green'])
plt.title("Average Coffee Price")
plt.xlabel('Coffee Type')
plt.ylabel('Average Price in $')
plt.show()
```

```python
commodity_prices.tail(4)
```

```python
# Ploting an Histogram

plt.figure(figsize=(8,5), dpi=300)
plt.hist(commodity_prices['sugar_world'], bins = 30, alpha=0.7, label="World Sugar Price", color='pink')
plt.hist(commodity_prices['sugar_us'], bins = 30, alpha=0.7, label="US Sugar Price", color='blue')
plt.hist(commodity_prices['sugar_eu'], bins = 30, alpha=0.7, label="The EU Sugar Price", color='green')
plt.title("Distribution of Sugar Prices")
plt.xlabel('Prices in $')
plt.ylabel('Frequency')
plt.show()
```

```python
commodity_prices.tail(4)
```

```python
# Scatter plots to check data against each other

plt.figure(figsize=(8,5), dpi=300)
plt.scatter(commodity_prices['coffee_arabica'], commodity_prices['sugar_world'], label='Arabica Coffee against Sugar Price', color='brown', alpha=0.5)
plt.scatter(commodity_prices['coffee_robustas'], commodity_prices['sugar_world'], label='Robustas Coffee against Sugar Price', color='yellow', alpha=0.5)
plt.title('Coffee Prices against Sugar Prices')
plt.xlabel("Coffee Price in $")
plt.ylabel("Sugar Price in $")
plt.legend(loc="upper right")
plt.show()
```
