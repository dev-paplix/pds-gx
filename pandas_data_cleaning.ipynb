# Data Cleaning Using Pandas

```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
# Importing the dataset
bios = pd.read_csv("./assets/new_bios.csv")
bios.head(4)
```

```python
bios.info()
```

```python
# Copying the data set to take care of null amounts
null_bios = bios.copy()
null_bios.info()
```

```python
# Removing all null data

null_bios = null_bios.dropna()
null_bios.info()
```

```python
null_bios.head()
```

```python
# Copying the dataset to take care of duplication
duplicate_bios = bios.copy()
duplicate_bios.info()
```

```python
# Removing duplication using Pandas
duplicate_bios = duplicate_bios.drop_duplicates()
duplicate_bios.info()
```

```python
bios.head()
```

```python
# Copy the dataset for removing dots
bios_nodot = bios.copy()
bios_nodot.head(5)
```

```python
# Creating a new column for name

bios_nodot['Name'] = bios_nodot['Used name'].str.replace('•', ' Bin ')
bios_nodot.head(5)
```

```python
# Making copy to split height and weight
hw_bios = bios.copy()
hw_bios.head(4)
```

```python
# Creating two new columns and split the Measurments

hw_bios[['Height', 'Weight']] = hw_bios['Measurements'].str.split('/', expand=True)
hw_bios.head(4)
```

```python
hw_bios.info()
```

```python
# Changing height to numeric

hw_bios['Height'] = pd.to_numeric(hw_bios['Height'].str.strip(' cm'), errors='coerce')

hw_bios.info()
```

```python
# Changing height to numeric

hw_bios['Weight'] = pd.to_numeric(hw_bios['Weight'].str.strip(' kg'), errors='coerce')

hw_bios.info()
```

```python
# Making height Desending
hw_bios.sort_values('Height', ascending=False)
hw_bios.head(4)
```

```python
# Making height Ascending
hw_bios.sort_values('Height', ascending=True)
hw_bios.head(4)
```

```python
# Changing the born date from objects to date
date_pattern = r'(\d+ \w+ \d{4}|\{4})'
hw_bios['born_date'] = hw_bios['Born'].str.extract(date_pattern)
hw_bios.info()
```

```python
hw_bios.head(4)
```

```python
# Extracting the born year

hw_bios['born_year'] = hw_bios['Born'].str.extract(r'(\d{4})')
hw_bios.head(4)
```

```python
# Converting the born date to date and time
hw_bios['born_date'] = pd.to_datetime(hw_bios['born_date'], format="mixed", errors='coerce')
hw_bios.info()
```

```python
hw_bios.head(4)
```

```python
# Converting born year to number
hw_bios['born_year'] = pd.to_numeric(hw_bios['born_year'])
hw_bios.info()
```
