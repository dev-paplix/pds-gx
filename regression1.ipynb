```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Create random for reproductivity
np.random.seed(42)
```

```python
# Generate synthetic data
# X values between -3 and 3
X = np.linspace(-3, 3, 100).reshape(-1, 1)
# Quadratic Relationp with noise: y= 2+3*x - 2*X^2 + noise
y = 2 + 3 * X.ravel() - 2 * X.ravel() ** 2 + np.random.normal(0, 2, X.shape[0])
```

```python
# Ploting the Synthetic data:

plt.scatter(X, y, color = 'blue', label = 'Data Points')
plt.xlabel('X')
plt.ylabel('y')
plt.title("Synthetic Data: Quadratic Relationship")
plt.legend(loc=("upper left"))
plt.show()
```

```python
# Fit a simple linear regression model
linear_model = LinearRegression()
linear_model.fit(X,y)
y_pred_linear = linear_model.predict(X)
```

```python
# Evaluation of the linear model

mse_linear = mean_squared_error(y, y_pred_linear)
print(f"Linear Regression MSE: {mse_linear:.2f}")
```

```python
# Plot the linear Regression line along with the data

plt.scatter(X, y, color = 'blue', label = 'Data Points')
plt.plot(X, y_pred_linear, color='red', label='Linear Regression', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.title("Linear Regression Fit")
plt.legend(loc=("upper left"))
plt.show()
```
